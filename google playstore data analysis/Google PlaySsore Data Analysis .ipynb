**Remove Duplicate data**

```python
google_playstore_df = google_playstore_df.drop_duplicates(subset=['App'],keep='first')
```

```python
google_playstore_df
```

```python
google_playstore_df.columns
```

```python
google_playstore_df.info()
```

**Converting the Reviews column to Numeric int64 as it is 'object' type and non numeric characters to 0.**

```python
pd.options.mode.chained_assignment = None 
```

```python
google_playstore_df['Reviews'] = pd.to_numeric(google_playstore_df.Reviews,errors='coerce').fillna(0).astype('int64')
```

```python
google_playstore_df['Reviews'] 
```

**Converting the Rating column to Numeric float64 as it is 'object' type and non numeric characters to 0.**

```python
google_playstore_df['Rating'] = pd.to_numeric(google_playstore_df.Rating,errors='coerce').fillna(0).astype('float64')
```

```python
google_playstore_df['Rating']
```

**removing + & , characters from Installs column and Converting to Numeric int64, non numeric characters to 0.**

```python
google_playstore_df['Installs'] = google_playstore_df['Installs'].str.replace('+','',regex=True)

google_playstore_df['Installs'] = google_playstore_df['Installs'].str.replace(',','',regex=True)

google_playstore_df['Installs'] = pd.to_numeric(google_playstore_df.Installs,errors='coerce').fillna(0).astype('int64')
```

```python
google_playstore_df['Installs'] 
```

**removing Dollar ($) character from Price column**

```python
google_playstore_df['Price'] = google_playstore_df['Price'].str.replace('$','',regex=True)
```

```python
google_playstore_df['Price'] 
```

**Replacing M with 000000 and k with 000 characters in Size column and Converting to Numeric int64, non numeric characters to 0**

```python
google_playstore_df['Size'] = google_playstore_df['Size'].str.replace('M','000000',regex=True)
google_playstore_df['Size'] = google_playstore_df['Size'].str.replace('k','000',regex=True)
google_playstore_df['Size'] = pd.to_numeric(google_playstore_df.Size,errors='coerce').fillna(0).astype('int64')
```

```python
google_playstore_df['Size']
```

**Replacing 0 in Type column to 'Free'**

```python
google_playstore_df['Type'] = google_playstore_df['Type'].str.replace('0','Free',regex=True)
```

```python
google_playstore_df['Type']
```

**Converting the Price column to Numeric float64 as it is 'object' type and non numeric characters to 0**

```python
google_playstore_df['Price'] = pd.to_numeric(google_playstore_df.Size,errors='coerce').fillna(0).astype('int64')
```

```python
google_playstore_df['Price']
```

# Q6 Give overview of Numerica data

```python
numeric_features = [feature for feature in google_playstore_df.columns if google_playstore_df[feature].dtype != 'O']
numeric_features
```

```python
google_playstore_df[numeric_features]
```

```python
plt.figure(figsize=(15, 15))
plt.suptitle('Univariate Analysis of Numerical Features', fontsize=20, fontweight='bold', alpha=0.8, y=1.)

for i in range(0, len(numeric_features)):
    plt.subplot(5, 3, i+1)
    sns.kdeplot(x=google_playstore_df[numeric_features[i]],shade=True, color='r')
    plt.xlabel(numeric_features[i])
    plt.tight_layout()
```
